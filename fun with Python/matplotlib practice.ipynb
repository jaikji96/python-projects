there are two types of multiplots

one can be created by simple methods and other is being created by the object oriented method.


lets discuss SIMPLE METHOD

```python
#syntax is plt.subplot(nrows,ncol,plot_number)
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

### matplotlib object oriented plotting

```python
fig= plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])# this will create a blank canvas.
#now plotting our plot
axes.plot(x,y)
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
```

```python
fig= plt.figure()
axes1=fig.add_axes([0.1,0.1,0.95,0.95])
axes2=fig.add_axes([0.25,0.7,0.5,0.2])
#now plotting our plot
axes1.plot(x,y)
axes1.set_xlabel('x axis')
axes1.set_ylabel('y axis')
axes1.set_title('plot1')

axes2.plot(y,x)
axes2.set_xlabel('X AXIS')
axes2.set_ylabel('Y AXIS')
axes2.set_title('PLOT2')
```

#### LEGENDS

we can use label text to acually what plot is what plot 

```python
fig4=plt.figure()
axes4=fig4.add_axes([0,0,1,1])
axes4.set_title('TITLE')
axes4.plot(x,x**2,label='X square')
axes4.plot(x,x**3,label='X cubed')
axes4.legend(loc=0)#loc parameter use to specify the position of label
```

#### setting a color

```python
a=plt.figure(figsize=(4,4))
b=a.add_axes([0,0,1,1])
b.plot(x,y,color='green')
#we can also pass a rgb color code and rgb hex code 
b.plot(y,x,color='#FF8C00')
```

```python
a,b=plt.subplots(figsize=(20,20))
b.plot(x,x+1,color='blue',alpha=0.5)
b.plot(x,x+2,color='purple')
b.plot(x,x+3,color='#FF8C00')
b.plot(x,x+4,color='red',linewidth=3) #makes line darker.
b.plot(x,x+5,color='green',lw=3,linestyle='--')
b.plot(x,x+6,color='green',lw=1,ls=':')
b.plot(x,x+7,color='yellow',lw=2,ls='-.')
b.plot(x,x+8,color='black',lw=2)
b.plot(x,x+9,color='grey',lw=1.75,ls='-',marker='+')
b.plot(x,x+10,color='pink',lw=1,ls='--',marker='o')
b.plot(x,x+11,color='pink',lw=2,ls='--',marker='s')#square marker
b.plot(x,x+12,color='blue',lw=2,ls='--',marker='1')
b.plot(x,x+13,color='magenta',lw=3,ls='dotted',marker='o',markersize=5)
b.plot(x,x+14,color='green',lw=1,ls='dashdot',marker='s',markersize=8)
b.plot(x,x+15,color='red',lw=3,ls='dashed',marker='o',markersize=10,markerfacecolor='green')
b.plot(x,x+16,color='black',lw=3,ls='dashdot',marker='s',markersize=13,markerfacecolor='yellow',markeredgewidth=4,markeredgecolor='red')
```
