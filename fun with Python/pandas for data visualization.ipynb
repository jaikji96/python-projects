# pandas for data visualisation

As we know that Data visualization are actually bulit off matplotlib but they allow you to call directly for data frame

```python
import numpy as np
import pandas as pd
%matplotlib inline

import seaborn as sns
```

```python
df1=pd.read_csv('df1',index_col=0)
df1
```

```python
df2=pd.read_csv('df2')
df2
```

##### let we want to get a histogram of all the values in the column A for df1 and with pandas we can do it correctly.

```python
df1['A'].hist()
```

```python
df1['A'].hist(bins=20,figsize=(10,10))
```

### NOTE:- if we do not run above hist command without importing seaborn and run the following command with importing seaborn then we will get a simple output without getting grid in our plot

```python
#using another method to get a histogram.
df1['A'].plot(kind='hist',bins=20)
```

```python
df2['a'].plot(kind='bar')
```

```python
df2.plot.area()
```

```python
df2.plot.area(alpha=0.5)
```

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

```python
df1['A'].plot.hist(bins=50)
```

```python
df1.plot.scatter(x='A',y='B')
```

```python
df1.plot.scatter(x='A',y='B',c='C')
```

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

```python
df1.plot.scatter(x='A',y='B',c='C',s=df1['C']*100,cmap='winter')
```

## using a box plot

```python
df2.plot.box(cmap='summer')
#taking every column as category and plot a box plot for us
```

```python
# hexagonal plot
df=pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
df.head()
```

```python
df.plot.hexbin(x='a',y='b',gridsize=20,cmap='coolwarm')
```
